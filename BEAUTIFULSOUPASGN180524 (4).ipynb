Ques 6:Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloadedarticles/
and scrapa)
Paper title
b) date
c) Author

```python
page6=requests.get('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')
page6
```

```python
soup6=BeautifulSoup(page6.content)
print(soup6.prettify())
```

```python
paper_title=[]
for pt in soup6.find_all('h2',class_="h5 article-title"):
    paper_title.append(pt.text.replace('\n',''))
    



date=[]
for dt in soup6.find_all('p',class_="article-date"):
    date.append(dt.text)



author=[]
for at in soup6.find_all('p',class_="article-authors"):
    author.append(at.text)
```
